```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

```

```python
image_path = os.path.join("Source", "scroll_horizontal_brown_basic.png")
image_path
```

```python
 # 读取图像

img = cv2.imread(image_path)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)

```

```python
# 预处理：降噪
gray = cv2.GaussianBlur(gray, (5, 5), 0)

plt.imshow(gray)

```

```python
# 边缘检测
edges = cv2.Canny(gray, 195, 200)
plt.imshow(edges)
```

```python
# 形态学操作增强边缘
kernel = np.ones((3, 3), np.uint8)
edges = cv2.dilate(edges, kernel, iterations=1)
plt.imshow(edges)
```
